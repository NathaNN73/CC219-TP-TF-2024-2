```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df50 = pd.read_csv('film_reviews_result.csv', sep='|')
df50.head(10)
```

```python
df50.info()
```

```python
df50_1 = df50.copy()
df50_1['film_avg_rate'] = df50_1['film_avg_rate'].str.replace(',', '.').astype(float)
df50_1['gender'] = df50_1['gender'].apply(lambda x: x.split(', '))
```

```python
df50_1.head(5)
```

```python
df50_1.info()
```

```python
df50_1.describe()
```

```python
df1000 = pd.read_csv('reviews_filmaffinity.csv', sep='~')
df1000.head(10)
```

```python
df1000.info()
```

```python
df1000.dropna(inplace=True)
df1000.info()
```

```python
df1000_1 = df1000.copy()
df1000_1['film_avg_rate'] = df1000_1['film_avg_rate'].str.replace(',', '.').astype(float)
df1000_1['review_rate'] = df1000_1['review_rate'].astype(int)
df1000_1['gender'] = df1000_1['gender'].apply(lambda x: [x])
```

```python
df1000_1.head(5)
```

```python
full_df = pd.concat([df50_1, df1000_1], ignore_index=True)
```

```python
full_df.info()
```

```python
full_df.describe()
```

```python
plt.figure(figsize=(18, 6), dpi=80)
sns.boxplot(data=full_df, orient="h",
            palette="Set1")
plt.title('Diagrama de cajas de las calificaciones de las películas')
plt.show()
```

```python
full_df.hist(figsize=(10,4))
plt.show()
```

```python
no_text_df = full_df.drop(['review_text', 'review_title'], axis=1)
no_text_df = no_text_df.explode('gender')
no_text_df
```

```python
gender_count = no_text_df.groupby('gender').count()
data = gender_count['film_name'].sort_values(ascending=False).head(20)

plt.figure(figsize=(10, 6), dpi=80)
plt.bar(data.index, data.values)
plt.title('Número de críticas por género (Top 20)')
plt.xticks(rotation=90)
plt.show()
```

```python
gender_mean = no_text_df.drop('film_name', axis=1).groupby('gender').mean()
data = gender_mean['review_rate'].sort_values(ascending=False).head(20)

plt.figure(figsize=(10, 6), dpi=80)
plt.bar(data.index, data.values)
plt.title('Media de las calificaciones por género (Top 20)')
plt.xticks(rotation=90)
plt.show()
```

```python
avg_ratings = full_df.groupby('film_name')['review_rate'].mean()

data = avg_ratings.sort_values(ascending=False).head(20)

plt.figure(figsize=(10, 6), dpi=80)
plt.bar(data.index, data.values)
plt.title('Media de las calificaciones por titulo (Top 20)')
plt.xticks(rotation=90)
plt.show()
```

```python
review_counts = full_df.groupby('film_name').size()
avg_review_scores = full_df.groupby('film_name')['review_rate'].mean()

plt.figure(figsize=(10, 6), dpi=80)
plt.scatter(review_counts, avg_review_scores, alpha=0.5)
plt.title('Numero de reviews vs Puntaje medio de reviews por Titulo')
plt.xlabel('Numero de Reviews')
plt.ylabel('Puntaje Medio')
plt.grid(True)
plt.show()
```
